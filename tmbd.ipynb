```python
data=data.drop('homepage', axis=1)
```

```python
data["popularity"] = pd.to_numeric(data["popularity"], errors='coerce')
```

```python
data["budget"] = pd.to_numeric(data["budget"], errors='coerce')
```

```python
data["vote_count"] = data["vote_count"].fillna(0).astype(int)
```

```python
data["id"] = data["id"].astype(str).fillna(0)
```

```python
data["revenue"] = pd.to_numeric(data["revenue"], errors='coerce')
```

```python
data["revenue"] = data["revenue"].fillna(0).astype(int)
data["runtime"] = data["runtime"].fillna(0).astype(int)
```

```python
data["revenue_adj"] = data["revenue_adj"].fillna(0).astype(int)
```

```python
data["budget_adj"] = data["budget_adj"].fillna(0).astype(int)
```

```python
data["budget"] = data["budget"].fillna(0).astype(int)
```

```python
data.info()
```

### Which genres are the most popular?

```python
most_popular_genres=data.groupby('genres')['popularity'].sum().sort_values(ascending=False).head(10)
most_popular_genres
```

```python
plt.figure(figsize=(12,10))
most_popular_genres.plot(kind='bar',color='skyblue',)

plt.title('most_popular_genres')
plt.xlabel('genres')
plt.ylabel('popularity')
```

### What is the trend of movie releases over the years?

```python
movie_trend= data.groupby('release_year')['imdb_id'].count().sort_values(ascending=False).head(10)
movie_trend
```

```python
plt.figure(figsize=(12,10))
movie_trend.plot(kind='barh',color='skyblue',)

plt.title('movie_trend')
plt.xlabel('id')
plt.ylabel('release_year')
```

### What directors have the highest average movie rating

```python
most_rated_directors = data.groupby('director')['vote_average'].sum().sort_values(ascending=False).head(10)
most_rated_directors
```

```python
plt.figure(figsize=(12,10))
most_rated_directors.plot(kind='barh',color='skyblue',)

plt.title('most_rated_directors')
plt.xlabel('vote_average')
plt.ylabel('director')
```

### Is there a relationship between budget and revenue

```python
correlation = data['budget_adj'].corr(data['revenue_adj']).round(2)
"moderate correlation",correlation
```

### Who are the most frequently featured actors

### What is the average runtime of movies by genre

```python
avg_runtime=data.groupby('genres')['runtime'].mean().sort_values(ascending=False).head(10)
avg_runtime
```

```python
plt.figure(figsize=(12,10))
avg_runtime.plot(kind='bar',color='skyblue',)

plt.title('avg_runtime')
plt.xlabel('genres')
plt.ylabel('runtime')
```
